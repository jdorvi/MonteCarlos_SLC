```python
import html5lib
from bs4 import BeautifulSoup
import requests
```

```python
transects = ['M10']#+str(i+17) for i in range(26)]
data_sources = "http://dune.seagrant.sunysb.edu/nyshore/profile_viewer/index.jsp?&currentProfile={}"
```

```python
NAME = ['survey=', '&current']
DIMS = ["Distance:", ", Elevation:", "', WIDTH"]
```

```python
for transect in transects:
    data_source = data_sources.format(transect)
    r = requests.get(data_source)
    soup = BeautifulSoup(r.text,'html5lib')
    r.connection.close()
    r.close()
    areas = soup.find_all(shape="POLY")
    outfile = transect+".txt"
    with open(outfile, 'w') as f:
        for area in areas:
            name_idx1 = area.attrs['href'].find(NAME[0])
            name_idx2 = area.attrs['href'].find(NAME[1])
            dims_idx1 = area.attrs['onmouseover'].find(DIMS[0])
            dims_idx2 = area.attrs['onmouseover'].find(DIMS[1])
            dims_idx3 = area.attrs['onmouseover'].find(DIMS[2])

            transect = area.attrs['href'][name_idx1+len(NAME[0]):name_idx2]
            distance = area.attrs['onmouseover'][dims_idx1+len(DIMS[0]):dims_idx2]
            elevation = area.attrs['onmouseover'][dims_idx2+len(DIMS[1]):dims_idx3]

            line = '{:<15}{:>12}{:>12}\n'.format(transect,distance,elevation)
            f.write(line)
    f.close()
```
